```python
an=sns.countplot(x='Gender',data=dat,width=0.3)
an
plt.show()
```

```python
an=sns.countplot(x='Gender',data=dat,width=0.3)

for bars in an.containers:
    an.bar_label(bars)
    
```

```python
an=sns.barplot(y='Monthly Income',x='CF_age band',data=dat,hue='Gender',palette='rainbow')
sns.set(rc={'figure.figsize':(6,20)})
sns.despine()
for bars in an.containers:
    an.bar_label(bars)
```

```python
eml_dep= dat.groupby(['CF_attrition label'],as_index=False)['Employee Number'].sum().sort_values(by='Employee Number',ascending=False)
eml_dep
```

```python
colr = sns.color_palette('terrain')[0:2]
plt.pie('Employee Number',labels='CF_attrition label',radius=0.5,colors=colr,autopct="%0.1f%%",data=eml_dep,startangle=75,shadow=True,explode=[0,0.2])
plt.show()
```

```python
trav=dat.groupby(['Business Travel'],as_index=False)['Employee Number'].sum().sort_values(by='Employee Number',ascending=False)
trav
```

```python
sns.barplot(x='Business Travel',y='Employee Number',data=trav,color='blue',palette='hls',width=0.5)
plt.xlabel("Business Travel")
plt.ylabel("Employees Number(cr)")
plt.show()
```

```python
sns.scatterplot(x="Job Role",y="Total Working Years",data=dat,hue="Gender",style="Gender")
plt.show()
width=20
height=7
sns.set(rc={'figure.figsize':(width,height)})
```

```python
sns.displot(dat,kind="kde",x="Percent Salary Hike",y="Work Life Balance")
plt.show()
```
